```python
from Zandpack.TimedependentTransport import TD_Transport as TDT
import numpy as np
from pickle import load
t1,t2 = 7,10
dk = 500.0
NCPU = 4
```

```python
Dev = load(open('Device.SiP','rb'))
gd  = Dev.to_sisl()
e1 = Dev.elecs[0]
e2 = Dev.elecs[1]
#e3 = Dev.elecs[2]


gem = e1.to_sisl()
gep = e2.to_sisl()
#ger = e3.to_sisl()
```

```python

R    = TDT([gem,gep], gd, kT_i=[0.025, 0.025], mu_i = [0.0, 0.0])
line = np.array([np.linspace(-5,5,121) + 1j * 0.01]*2)
R.Make_Contour(line, 17,pole_mode='JieHu2011')
Contour = R.Contour

Dev.run_TS_RSSE_calc((t1,t2,1), Contour, dk = dk,dk_fail = 10.0,
                     parallel_E = True, num_procs = NCPU,fix_dir = 'C', 
                     DOS_GF=True,reuse_fdf=[False,True] 
                     # reuse_fdf is for keeping the synthetic atoms block in the RSSE calculation
)



```

```python
R.Device = Dev
```

```python
sub_idx = []
cC,cAu  = 4, 6
count = 0
for ia in range(len(Dev.s)):
    if Dev.s[ia]==6:
        sub_idx += [count + 2]
        count   += cC
    if Dev.s[ia] == 79:
        sub_idx += [count, count+1,count+2, count+3, count+4, count+5]
        count   += cAu

R.read_data(sub_orbital=sub_idx)
```
